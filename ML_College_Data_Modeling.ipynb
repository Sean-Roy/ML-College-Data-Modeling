```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
# Just a quick summary of the cluster centers as per the model's analysis:
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# Since we have access to the True values (normally not the case), we can test to see the accuracy:
df.info()
```

```python
df['Private'].unique()
```

```python
# Changing the object type to a binary value for evaluation (since we have actual results for comparison):
def group(private):
    if private == 'Yes':
        return 1
    else:
        return 0

df['Group'] = df['Private'].apply(group)
df['Group']
```

```python
# Check the confusion matrix and classification report:
print(f'Confusion matrix:\n\n{confusion_matrix(df['Group'],kmeans.labels_)}\n')
print(f'Classification report:\n\n{classification_report(df['Group'],kmeans.labels_)}')
```
